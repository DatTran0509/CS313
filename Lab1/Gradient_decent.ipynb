```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
# Đọc dữ liệu từ file CSV
df = pd.read_csv('./ParisHousing.csv')
```

```python
df.head(10)
```

```python
# Chọn các cột đầu vào và cột mục tiêu
X = df.drop(columns=['price'])  # Loại bỏ cột giá
y = df['price']

# Chuẩn hóa dữ liệu (scaling) cho các đặc trưng
X = (X - X.mean()) / X.std()

```

```python
# Thêm một cột 1 vào X để tương ứng với hệ số b trong phương trình hồi quy (bias term)
X = np.c_[np.ones(X.shape[0]), X]

# Chuyển dữ liệu thành dạng numpy array
X = np.array(X)
y = np.array(y)
```

```python
# Khởi tạo các tham số của mô hình (trọng số ngẫu nhiên)
theta = np.random.randn(X.shape[1])

# Thiết lập các thông số cho Gradient Descent
learning_rate = 0.01
iterations = 1000
m = len(y)

# Hàm tính toán giá trị hàm mất mát (cost function)
def compute_cost(X, y, theta):
    predictions = X.dot(theta)
    cost = (1/(2*m)) * np.sum((predictions - y) ** 2)
    return cost

# Hàm cập nhật trọng số bằng Gradient Descent
def gradient_descent(X, y, theta, learning_rate, iterations):
    cost_history = np.zeros(iterations)
    
    for i in range(iterations):
        gradients = (1/m) * X.T.dot(X.dot(theta) - y)
        theta -= learning_rate * gradients
        cost_history[i] = compute_cost(X, y, theta)
    
    return theta, cost_history
```

```python
# Chạy Gradient Descent để tìm các tham số tối ưu
theta_optimal, cost_history = gradient_descent(X, y, theta, learning_rate, iterations)
```

```python
# In ra kết quả
print("Các tham số của mô hình tối ưu (theta):", theta_optimal)

# Dự đoán giá nhà với các tham số tối ưu
predictions = X.dot(theta_optimal)
```

```python
# In kết quả dự đoán
print("Dự đoán giá nhà:", predictions[:5])  # In ra 5 dự đoán đầu tiên
```
